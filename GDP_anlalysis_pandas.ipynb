```python
# import GSVA FILES TO MERGE WITH STATES
import os
```

```python
gsva=pd.DataFrame()
```

```python
path="E:/DS/gvbcmyprojects/03_ GDP_ Niti Ayog/Datasets/GSVA/"
gsv_all=os.listdir(path)
```

```python
print(len(gsv_all))
print(len(states))
```

```python
for filename in gsv_all:
    for statename in states:
        if statename in filename:
            filepath=path+filename
            data=pd.read_csv(filepath,encoding="ISO-8859-1")
            data=data[['Item','2014-15']]
            data=data.T
            data.columns=data.iloc[0]
            data=data[1:]
            data["state"]=statename
            gsva=pd.concat([data,gsva])         
            
```
